## Solution
1. find the midle element of the list
2. if it matches the queries no. return midle position as ans
3. if it is less than queried no, move left
4. if it is greater than queries no, move right
5. if no elements remain return -1 


```python
def test_location(cards, query, mid):
    mid_number = cards[mid]
    if mid_number == query:
        if mid-1 >= 0 and cards[mid-1] == query:
            return 'left'
        else:
            return 'found'
    elif mid_number < query:
        return 'left'
    else:
        return 'right'

def locate_card(cards, query):
    low, high = 0, len(cards)-1
    while low <= high:
        mid = (low+high)//2
        result = test_location(cards,query,mid)

        if result == 'found':
            return mid
        elif result == 'left':
            high = mid - 1
        else:
            low = mid + 1
    return -1
```
